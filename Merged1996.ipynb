```python
import pandas as pd
import numpy as np
```

```python
transaction=pd.read_csv("transaction1.csv")
stores=pd.read_csv("stores1.csv")
period=pd.read_csv("period1.csv")
sku=pd.read_csv("sku1.csv")
region=pd.read_csv("region1.csv")
salesman=pd.read_csv("salesman1.csv")
transaction.head()
```

```python
period.rename(columns={"PeriodID":"Period","Period.1":"PeriodID"},inplace=True)
period.head()
```

```python
merge1=pd.merge(transaction,stores, on="StoreID",how="outer")
```

```python
merge2=pd.merge(merge1,period,on="PeriodID",how="outer")
```

```python
merge3=pd.merge(merge2,sku,on="SKUCode",how="outer")
```

```python
region.head(1)
```

```python
merge4=pd.merge(merge3,salesman,on="SalesmanID",how="outer")
```
